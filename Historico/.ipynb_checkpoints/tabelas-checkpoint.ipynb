```python
import pandas as pd

df = pd.read_excel('base dash.xlsx', sheet_name='Planilha1')
df.head()
```

```python
df.columns
```

```python
for i in df[['Opióide de resgate na RPA (sim ou não)','Uso de opióide prescrito de horário nas primeiras 24 horas',
            'uso de opióide nas primeiras 36 horas','Dor com escala verbal numérica maior que maior que 3 na internação (sim ou não)',
            'Uso de opióide de resgate nas primeiras 24 horas (sim ou não)','Uso de opióide de resgate nas primeiras 24 horas (sim ou não)',
             'Uso de opioide de resgate na primeiras 36 horas (sim ou não)']]:
    a = df.groupby([i])[i].count().reset_index(name='fi')
    a['fr'] = round((a['fi']*100)/sum(a['fi']),2)
    display(a)
```

```python
for i in df[['Opióide de resgate na RPA (sim ou não)','Uso de opióide prescrito de horário nas primeiras 24 horas',
            'uso de opióide nas primeiras 36 horas','Dor com escala verbal numérica maior que maior que 3 na internação (sim ou não)',
            'Uso de opióide de resgate nas primeiras 24 horas (sim ou não)','Uso de opióide de resgate nas primeiras 24 horas (sim ou não)',
             'Uso de opioide de resgate na primeiras 36 horas (sim ou não)']]:
    a = df.groupby([i, 'Realização de bloqueio do plano eretor da espinha (sim ou não)']).size().unstack()
    for t in a.columns:
        a[t].fillna(0, inplace=True)
        a[t] = a[t].apply(lambda x: int(x))
    a.loc['total'] = a.sum(axis=0)
    b = a.copy()
    for c in a.columns:
        a[c] = a[c]*100/a[c]['total']
    display(round(a, 2))
    display(round(b, 2))
    print("======"*15)
```
